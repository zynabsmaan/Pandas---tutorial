**18 - Split string**
    <div>We can split the string column to two or multiples columns</div>

```python
# df.Name.str.split(spliter, n, expand)
# spliter : the eapression you want to split to 
# n : number of limit splits but(0, -1, None) --> returns all splits 
# expand : false --> returns the list of splits 
#          True  --> returns the dimention of splits 


df.Name.str.split(" ", n=1, expand=True)
```

**To add the new spliting columns to the dataframe ** 

```python
# 
# First definie to how many number of columns you want to split to 
# Say you want to split column "Name" to "first" column and "last" column
# The command below shows how yu can implement it 

df['first_1'], df['last_1'] = df.Name.str.split(" ", n=1, expand=False).str

# Or 

df['first_2'], df['last_2'] = df.Name.str.split(" ", n=1, expand=True)[0], df.Name.str.split(" ", n=1, expand=True)[1]

df.head()
```

<h3 style="color:blue">18 - if else</h3>

<li>We can convert normal if else in python with "np.where" for two conditions or "np.select" from numpu library</li>
<div>For example</div>

```python
# Assume adding new column if a condition is trus --> yese, false--> no
def name(row):
    if row["Pclass"] in [1, 2]:
        return "yes"
    else:
        return "no"

df['flag'] = df.apply(name, axis=1)
df.head()
```

<br>
<br>
**23 - np.where**

```python
# Using by np.where
# First import package

import numpy as np 

df["flag2"] = np.where(df['Pclass'].isin([1, 2]), "yes", "no")
df.head()
```

<br>
<br>
**24 - np.select**

```python
# By using np.select 
# first set your conditions first in list, consider it as the statement after "if"
conditions = [df['Pclass'] == 1, df['Pclass'] == 2, df['Pclass'] == 3]

# consider it as "then" in "if then"
choices = ['yes', 'no', 'okay']

df["flag3"] = np.select(conditions, choices, default="noAction")
df.head()
```

<h3 style="color:blue">tqdm</h3>

<li>with very large datasets when using "map", "apply" and "applymap" pandas takes time 
    So by using tqdm, it predicts the time to finish the operations on datasets.</li>

```python
# first import the package 
from tqdm import tqdm_notebook
tqdm_notebook().pandas()
```

```python
# We change the map      --> progress_map 
#               apply    --> progress_apply
#               applymap --> progress_applymap

df['Name'].progress_map(lambda x: "Zynab Smaan")
```
